```python
import spacy
import re
import pandas as pd
from nltk.stem import PorterStemmer
```

```python
nlp = spacy.load("en_core_web_lg")
```

```python
def filter_keywords_by_score(keyword_list, threshold=0.4):
    filtered_keywords = []

    for item in keyword_list:
        # Split the item by whitespace and extract the keyword and score
        elements = item.strip().split()
        
        # Get the last element of the list, which contains the score
        score_str = elements[-1]

        # Try converting the score to float
        try:
            score = float(score_str)
        except ValueError:
            # If the conversion fails, skip this item and continue to the next one
            continue
        
        keyword = " ".join(elements[:-1])  # Join the remaining elements as the keyword

        if score > threshold:
            filtered_keywords.append(keyword)

    return filtered_keywords
```

```python
def returnLines(x):
    with open(x) as f:
        lines = f.readlines()
    return lines
```

```python
def stem_word_list(word_list):
    stemmer = PorterStemmer()
    stemmed_list = [stemmer.stem(word) for word in word_list]
    return stemmed_list
```

```python
print("Videopoints Keywords\n")
videopointsKey=[]
count = 1
while (count < 16):
    print('Section '+ str(count))
    videopointsKey.append(filter_keywords_by_score(returnLines('sec' + str(count)), threshold=0.4))
    print(filter_keywords_by_score(returnLines('sec' + str(count)), threshold=0.4))
    count = count + 1
    print("\n")
```

```python
def calculate_confusion_matrix(predicted_keywords, ground_truth_keywords, text_lengths):
    
    tp = 0  # True Positives
    fp = 0  # False Positives
    fn = 0  # False Negatives
    tn = 0  # True Negatives

    for i in range(len(predicted_keywords)):
        predicted = set(predicted_keywords[i])
        ground_truth = set(ground_truth_keywords[i])

        # Calculate True Positives
        tp += len(predicted.intersection(ground_truth))

        # Calculate False Positives
        fp += len(predicted.difference(ground_truth))

        # Calculate False Negatives
        fn += len(ground_truth.difference(predicted))

        # Calculate True Negatives
        tn += text_lengths[i] - len(ground_truth.union(predicted))

    return tp, fp, fn, tn
```
