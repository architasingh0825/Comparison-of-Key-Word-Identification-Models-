```python
def stem_word_list(word_list):
    stemmer = PorterStemmer()
    stemmed_list = [stemmer.stem(word) for word in word_list]
    return stemmed_list
```

```python
def extract_top_keywords(text, num_one_word, num_two_word, num_three_word):
    model = SentenceTransformer("paraphrase-MiniLM-L6-v2")
    kw_extractor = KeyBERT(model=model)

    # Tokenize text and remove punctuation
    tokens = word_tokenize(text)
    tokens = [word.lower() for word in tokens if word.isalnum()]

    # Remove stopwords
    stop_words = set(stopwords.words("english"))
    tokens = [word for word in tokens if word not in stop_words]

    # Reconstruct cleaned text
    cleaned_text = " ".join(tokens)

    # Specify the BioBERT model name
    bio_bert_model_name = "bionlp/bluebert_pubmed_mimic_uncased_L-12_H-768_A-12"

    # Load the BioBERT tokenizer and model
    tokenizer = AutoTokenizer.from_pretrained(bio_bert_model_name)
    model = AutoModel.from_pretrained(bio_bert_model_name)
    
    top_one_word = kw_extractor.extract_keywords(
        cleaned_text,
        keyphrase_ngram_range=(1, 1),
        top_n=num_one_word,
        stop_words="english",
    )
    
    top_two_word = kw_extractor.extract_keywords(
        cleaned_text,
        keyphrase_ngram_range=(1, 2),
        top_n=num_two_word,
        stop_words="english",
    )
    
    top_three_word = kw_extractor.extract_keywords(
        cleaned_text,
        keyphrase_ngram_range=(1, 3),
        top_n=num_three_word,
        stop_words="english",
    )
    
    # Extract only the keywords without the similarity scores
    top_one_word = [keyword for keyword, score in top_one_word]
    top_two_word = [keyword for keyword, score in top_two_word]
    top_three_word = [keyword for keyword, score in top_three_word]

    return top_one_word, top_two_word, top_three_word
```

```python
def calculate_confusion_matrix(predicted_keywords, ground_truth_keywords, text_lengths):
    
    tp = 0  # True Positives
    fp = 0  # False Positives
    fn = 0  # False Negatives
    tn = 0  # True Negatives

    for i in range(len(predicted_keywords)):
        predicted = set(predicted_keywords[i])
        ground_truth = set(ground_truth_keywords[i])

        # Calculate True Positives
        tp += len(predicted.intersection(ground_truth))

        # Calculate False Positives
        fp += len(predicted.difference(ground_truth))

        # Calculate False Negatives
        fn += len(ground_truth.difference(predicted))

        # Calculate True Negatives
        tn += text_lengths[i] - len(ground_truth.union(predicted))

    return tp, fp, fn, tn
```

```python
def calculate_metrics(tp, tn, fp, fn):
    # Calculate precision
    precision = tp / (tp + fp) if (tp + fp) > 0 else 0

    # Calculate recall (sensitivity)
    recall = tp / (tp + fn) if (tp + fn) > 0 else 0

    # Calculate F1-score
    f1_score = 2 * (precision * recall) / (precision + recall) if (precision + recall) > 0 else 0

    # Calculate accuracy
    accuracy = (tp + tn) / (tp + tn + fp + fn) if (tp + tn + fp + fn) > 0 else 0

    return precision, recall, f1_score, accuracy
```
