```python
def calculate_confusion_matrix(predicted_keywords, ground_truth_keywords, text_lengths):
    
    tp = 0  # True Positives
    fp = 0  # False Positives
    fn = 0  # False Negatives
    tn = 0  # True Negatives

    for i in range(len(predicted_keywords)):
        predicted = set(predicted_keywords[i])
        ground_truth = set(ground_truth_keywords[i])

        # Calculate True Positives
        tp += len(predicted.intersection(ground_truth))

        # Calculate False Positives
        fp += len(predicted.difference(ground_truth))

        # Calculate False Negatives
        fn += len(ground_truth.difference(predicted))

        # Calculate True Negatives
        tn += text_lengths[i] - len(ground_truth.union(predicted))

    return tp, fp, fn, tn
```
